# 删除无效的括号

## [题目](https://leetcode.cn/problems/remove-invalid-parentheses/description/)

- 给你一个由若干括号和字母组成的字符串 s ，删除最小数量的无效括号，使得输入的字符串有效。

- 返回所有可能的结果。答案可以按 任意顺序 返回。

- 示例1：
```C++
输入：s = "()())()"
输出：["(())()","()()()"]
```

- 示例2：
```C++
输入：s = "(a)())()"
输出：["(a())()","(a)()()"]
```

- 示例3：
```C++
输入：s = ")("
输出：[""]
```

- 提示：
    - 1 <= s.length <= 25
    - s 由小写英文字母以及括号 '(' 和 ')' 组成
    - s 中至多含 20 个括号

## 代码示例
- 回溯算法 + 剪枝
    - 时间复杂度：$O(n \times 2^n)$
    - 空间复杂度：$O(n^2)$

````python
class Solution {
public:
    vector<string> res;

    vector<string> removeInvalidParentheses(string s) {
        int lremove = 0;
        int rremove = 0;
    
        // 遍历字符串，计算要删除的左括号和右括号数量
        for (char c : s) {
            if (c == '(') lremove++;
            else if (c == ')') {
                if (lremove == 0) rremove++;
                else lremove--;
            }
        }
        // 使用递归实现回溯 + 剪枝
        helper(s, 0, lremove, rremove);
        return res;        
    }

    void helper(string str, int start, int lremove, int rremove) {
        // 终止条件
        if (lremove == 0 && rremove == 0) {
            if (isValid(str)) res.push_back(str);
            return ;
        }
        for (int i = start; i < str.size(); i++) {
            // 如果字符连续一致相等
            if (i != start && str[i] == str[i - 1]) continue;
            // 如果剩余字符无法满足去掉的数量要求，直接返回
            if (lremove + rremove > str.size() - i) return;
            // 尝试去掉一个左括号
            if (lremove > 0 && str[i] == '(') {
                helper(str.substr(0, i) + str.substr(i + 1), i, lremove - 1, rremove);
            }
            // 尝试去掉一个右括号
            if (rremove > 0 && str[i] == ')') {
                helper(str.substr(0, i) + str.substr(i + 1), i, lremove, rremove - 1);
            }
        }
    }

    bool isValid(const string& str) {
        // 遍历字符串查看当前括号规则是否匹配
        int cnt = 0;
        for (int i = 0; i < str.size(); i++) {
            if (str[i] == '(') cnt++;
            else if (str[i] == ')') {
                cnt--;
                if (cnt < 0) return false;
            }
        }
        return cnt == 0;
    }
};
````
