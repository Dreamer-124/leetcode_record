# 实现 strStr()

## [题目](https://leetcode.cn/problems/find-the-index-of-the-first-occurrence-in-a-string/)
- 给你两个字符串 haystack 和 needle ，请你在 haystack 字符串中找出 needle 字符串的第一个匹配项的下标（下标从 0 开始）。如果 needle 不是 haystack 的一部分，则返回  -1 

- 示例1：
```C++
输入：haystack = "sadbutsad", needle = "sad"
输出：0
解释："sad" 在下标 0 和 6 处匹配。第一个匹配项的下标是 0 ，所以返回 0
```

- 示例2：
```C++
输入：haystack = "leetcode", needle = "leeto"
输出：-1
解释："leeto" 没有在 "leetcode" 中出现，所以返回 -1 
```


- 提示
    - 1 <= haystack.length, needle.length <= 104
    - haystack 和 needle 仅由小写英文字符组成

## 代码示例
- 时间复杂度：$O(m + n)$
- 空间复杂度：$O(m)$，只需要保存字符串 needle 的前缀表

````python
class Solution {
public:
    // 获取到 next 数组
    void getNext(int* next, const string& s) {
        // next 数组的定义：next[i] 表示 s[0..i] 的最长相等前后缀的长度
        // 例如：s = "abcab" next[4] = 2
        int j = 0;  // 指向模式串起始位置
        next[0] = 0;  // 初始化起始的 next 数组
        // 这里的 i 是从 1 开始的，因为 next[0] 已经初始化了
        for(int i = 1; i < s.size(); i++) {
            // 这里的 j 是指向模式串的前缀
            while (j > 0 && s[i] != s[j]) {
                j = next[j - 1];
            }
            // 如果 s[i] 和 s[j] 相等，说明找到了一个相等的前后缀
            if (s[i] == s[j]) {
                j++;
            }
            next[i] = j;  // 这里的 next[i] 是指向模式串的后缀
        }
    }
    
    int strStr(string haystack, string needle) {
        // 如果 needle 为空字符串，返回 0
        if (needle.size() == 0) {
            return 0;
        }
        // 初始化 next 数组
        vector<int> next(needle.size());
        getNext(&next[0], needle);
        int j = 0;  // 指向模式串的起始位置
        // 遍历 haystack 字符串
        // 这里的 i 是指向 haystack 字符串的起始位置
        for (int i = 0; i < haystack.size(); i++) {
            // 如果 haystack[i] 和 needle[j] 不相等，说明需要跳过
            while(j > 0 && haystack[i] != needle[j]) {
                j = next[j - 1];
            }
            // 如果 haystack[i] 和 needle[j] 相等，说明找到了一个相等的前后缀
            if (haystack[i] == needle[j]) {
                j++;
            }
            // 如果 j 等于 needle 的长度，说明找到了匹配的子串
            if (j == needle.size() ) {
                return (i - needle.size() + 1);
            }
        }
        return -1;
    }
};
````
