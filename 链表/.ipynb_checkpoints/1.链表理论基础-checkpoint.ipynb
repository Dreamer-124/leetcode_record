# 链表理论基础
- **链表是一种通过指针串联在一起的线性结构**
- **单链表：每一个节点由两部分组成，一个是数据域，一个是指针域（存放指向下一个节点的指针），最后一个节点的指针域指向 null**
- **双链表：每一个节点有两个指针域，一个指向下一个节点，一个指向上一个节点（双链表既可以向前查询也可以向后查询）**
- **循环链表：链表的首尾相连（可以用来解决约瑟夫环问题）**
- **链表在内存中不是连续分布的，链表是通过指针域的指针链接在内存中各个节点，所以链表中的节点在内存中不是连续分布的，而是散乱分布在内存中的某地址上，分配机制取决于操作系统的内存管理**
- **链表的定义方式（C/C++）:**
    - **不定义构造函数使用默认构造函数的话，默认构造函数不会初始化任何成员变量，即在初始化的时候就不能直接给变量赋值**
```C++
// 单链表
struct ListNode {
    int val;  // 节点上存储的元素
    ListNode *next;  // 指向下一个节点的指针
    ListNode(int x) : val(x), next(NULL) {}  // 节点的构造函数
};

ListNode* head = new ListNode(5);  // 初始化节点
```
- **删除节点：改变节点的指针指向，C++中最好是再手动释放内存**
- **添加节点：改变节点的指针指向，添加数据域内容**
- **性能分析**
    - **数组适合数据量固定，频繁查询，较少增删的场景**
    - **链表适合数据量不固定，频繁增删，较少查询的场景**
